```python
#importing pandas library which has all the functions related to table calculations and renaming it
import pandas as pd
```

```python
#reading our dataset from Salary_Data.csv
dataset = pd.read_csv("Salary_Data.csv")
```

```python
#view the dataset which we imported
dataset
```

```python
#assinging our input to variable independent
independent = dataset[["YearsExperience"]]
independent
```

```python
#assigning the output variable to dependent
dependent = dataset[["Salary"]]
dependent
```

```python
#Spliting the training and test dataset using sklearn library
from sklearn.model_selection import train_test_split
#calling the Split function and adding the split percentage and also assigning it to variable
X_train,X_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#importing LinearRegression class from sklearn
from sklearn.linear_model import LinearRegression
#loading all the functions of LinearRegression
regressor=LinearRegression()
#giving the parameters ie the input to the model to train ie substituting the formula of Linear line using fit to calulate weight and bias
regressor.fit(X_train,y_train)
```

```python
#To see what weight has be calculated for the model
weight=regressor.coef_
weight
```

```python
#To see the bias 
bias=regressor.intercept_
bias
```

```python
#predict the test data
y_pred=regressor.predict(X_test)
y_pred
```

```python
#calculating the r2 score
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
#save the model
import pickle
#creating a filename
filename="finalised_model_linear.sav"
```

```python
#saving the model by writing it into our space
pickle.dump(regressor,open(filename,'wb'))
```

```python
#load the model and check the output
load_model=pickle.load(open("finalised_model_linear.sav",'rb'))
```

```python
#predict salary for 13 years of experience
result=load_model.predict([[13]])
```
