#IQR

```python
descriptive = pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%","99%","Q4:100%","IQR","IQR_Range","LesserOutRange","GreaterOutRange","Min","Max"],columns=quan)
for columnName in quan:
    descriptive[columnName]["Mean"]=dataset[columnName].mean()
    descriptive[columnName]["Median"]=dataset[columnName].median()
    descriptive[columnName]["Mode"]=dataset[columnName].mode()[0]
    descriptive[columnName]["Q1:25%"]=dataset.describe()[columnName]["25%"]
    descriptive[columnName]["Q2:50%"]=dataset.describe()[columnName]["50%"]
    descriptive[columnName]["Q3:75%"]=dataset.describe()[columnName]["75%"]
    descriptive[columnName]["99%"]=np.percentile(dataset[columnName],99)
    descriptive[columnName]["Q4:100%"]=dataset.describe()[columnName]["max"]
    descriptive[columnName]["IQR"]=descriptive[columnName]["Q3:75%"]-descriptive[columnName]["Q1:25%"]
    descriptive[columnName]["IQR_Range"]= 1.5*descriptive[columnName]["IQR"]
    descriptive[columnName]["LesserOutRange"]= descriptive[columnName]["Q1:25%"] - descriptive[columnName]["IQR_Range"]
    descriptive[columnName]["GreaterOutRange"]=descriptive[columnName]["Q3:75%"]+ descriptive[columnName]["IQR_Range"]
    descriptive[columnName]["Min"] = dataset[columnName].min()
    descriptive[columnName]["Max"] = dataset[columnName].max()
```

```python
descriptive
```

# Finding Outlier

```python
descriptive["hsc_p"]["Min"]<descriptive["hsc_p"]["LesserOutRange"]
```

```python
lesser=[]
greater=[]

for columnName in quan:
    if(descriptive[columnName]["Min"]<descriptive[columnName]["LesserOutRange"]):
       lesser.append(columnName)
    if(descriptive[columnName]["Max"]<descriptive[columnName]["GreaterOutRange"]):
        greater.append(columnName)
    
```

```python
for columnName in lesser:
    dataset[columnName][dataset[columnName]<descriptive["LesserOutRange"]] =descriptive[columnName]["LesserOutRange"]
for columnName in greater:
    dataset[columnName][dataset[columnName]>descriptive["GreaterOutRange"]] =descriptive[columnName]["GreaterOutRange"]
       
```
